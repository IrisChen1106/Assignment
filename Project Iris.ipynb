```python
def simulate_game(num_glasses):
    glasses = [0.0 for _ in range(num_glasses)]

    while True:
        total_pour_amount = 0.5  # Total amount Player 1 can pour
        print("\nPlayer 1's turn (total 0.5 pint to distribute):")

        while total_pour_amount > 0:
            print(f"Remaining liquid to distribute: {total_pour_amount}")
            pour_amount = float(input("Enter the amount of liquid to pour into a glass: "))
            if pour_amount > total_pour_amount:
                print("Cannot pour more than the remaining liquid.")
                continue
            
            glass_index = int(input(f"Choose the glass to pour into (0-{num_glasses-1}): "))
            if glass_index < 0 or glass_index >= num_glasses:
                print("Invalid glass number.")
                continue

            glasses[glass_index] += pour_amount
            total_pour_amount -= pour_amount

        print(f"Glasses after Player 1's turn: {glasses}")

        if any(glass > 1 for glass in glasses):
            print("Overflow detected! Player 1 wins!")
            break

        print("\nPlayer 2's turn:")
        glass_to_empty_1 = int(input(f"Choose the first of two adjacent glasses to empty (0-{num_glasses-1}): "))
        glass_to_empty_2 = (glass_to_empty_1 + 1) % num_glasses  # Automatically choose the adjacent glass

        glasses[glass_to_empty_1] = 0.0
        glasses[glass_to_empty_2] = 0.0

        print(f"Glasses after Player 2's turn: {glasses}")

        if all(glass <= 1 for glass in glasses):
            print("No overflow, Player 2 managed to prevent an overflow this round.")

# Start the game with a specific number of glasses
num_glasses = int(input("Enter the number of glasses: "))
simulate_game(num_glasses)
```
